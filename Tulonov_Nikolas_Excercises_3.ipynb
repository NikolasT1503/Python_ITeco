```
sentence = "А вы ноктюрн сыграть могли бы на флейте водосточных труб?"
words = sentence.split()
word_lengths = [len(word) for word in words if word != "the"]
print(words)
print(word_lengths)
```

Упражнение №8

Напишите программу для отрисовки такой красоты (паттерна) используя вложенный цикл.

image.png

```
x = int(input('Введите число: '))
for i in range(x*2-1):
    if i >= x:
        for j in range((x*2-1) - i):
            print('*', end='')
    else:
        for j in range(i+1):
            print('*', end='')
    print()
```

Упражнение №9

Пишем мини игру. Программа случайным образом выбирает число от 0 до 100.
Далее пользователь пытается это число угадать.

* Первый ввод - пользователю говорится холодно или горячо (считаем с шагом в 10 - если он написал 3 а загаданное машиной число 14, то ему будет сказано холодно)

* Последующие попытки - Пользователю говорится "Теплее" или "Холоднее" в зависимости от того ближе или дальше он от цели чем в предыдущий ход.

* Когда пользователь угадывает - завершаем программу, выводим поздравление на экран.



```
import random

a = random.randint(1, 1000)
print('Это смертельная игра!!! Вы должны угадать число, загаданное компьютером:')
while True:
    x = int(input('Вводи любое целое число: '))
    if abs(x-a) < 100:
        if abs(x-a) > 10:
            print('чуть потеплело, но все равно холодно')
        elif abs(x-a) < 10 and x != a:
            print('Теплее, осталось совсем чуть-чуть')
        else:
            print('Поздравляю, ты спасен!!!')
            break
    elif abs(x-a) > 100:
        print('Совсем холодно')
```
